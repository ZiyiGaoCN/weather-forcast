```python
import numpy as np
from yl import distribution,expect,get_param

x = [i for i in range(20)]

key = np.linspace(0, 10,40)
for i in range(len(key)):
    key[i] = get_param(key[i])
lines = []
for i in range(len(key)):
    lines.append(distribution(key[i]))

from matplotlib import pyplot as plt
import io
from PIL import Image
# plt.plot(x, line_0, label='0')
images = []
x = np.array(x)
x = x + 0.4
for line in lines:
    fig, ax = plt.subplots()
    ax.bar(x, line,width=0.8)
    ax.set_xlim(0, 20)
    ax.set_ylim(0, 0.4)
    # set x title of 'steps'
    # set y title of 'probability'
    # set title of 'distribution'
    ax.set_title('distribution')
    
    ax.set_xlabel('step')
    ax.set_ylabel('probability')

    buf = io.BytesIO()
    plt.savefig(buf, format='png')
    buf.seek(0)
    image = Image.open(buf)
    images.append(image)
    # buf.close()

    
images[0].save('distributions.gif', save_all=True, append_images=images[1:], optimize=False, duration=100, loop=0)
```
